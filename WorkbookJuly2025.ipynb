## Critical Zoom Analysis with Improved Parallelization

The improved critical zoom analysis now uses efficient single-level parallelization:

- **Top Level**: Parallelizes across temperatures (n = number of CPU cores)
- **Internal**: Forces sequential processing within each temperature calculation
- **Result**: Eliminates nested parallelization overhead and resource contention

This approach is more efficient than the previous nested parallelization where:
- 8 processes (temperatures)  8 processes (d0 values) = 64 competing processes on 8 CPUs
- Now: exactly 8 processes on 8 CPUs, each handling one temperature efficiently

```python
# Recreate and save the phase diagram plot for ml=0.1, lambda1=5.0
import numpy as np
import matplotlib.pyplot as plt
import os

# Parameters for this plot
mq = 0.1
lambda1_value = 5.0

print(f"Creating phase diagram plot for mq={mq}, lambda1={lambda1_value}")

# Load the improved phase diagram data (only the first 3 columns: mu, Tc, order)
phase_data = np.loadtxt('CP_data/phase_diagram_improved_ml0.1_lambda15.0.csv', delimiter=',', skiprows=1, usecols=(0, 1, 2))
mu_phase = phase_data[:, 0]
Tc_phase = phase_data[:, 1] 
order = phase_data[:, 2]  # Phase order (1 or 2)

# Load vector melting data
vector_data = np.loadtxt('vector_melting.csv', delimiter=',', skiprows=1)
mu_vector = vector_data[:, 0]
T_vector = vector_data[:, 1]

# Load axial melting data for mq=0.1, lambda1=5.0
axial_data = np.loadtxt('axial_melting_data_mq0.1_lambda5.0.csv', delimiter=',', skiprows=4)
mu_axial = axial_data[:, 0]
T_axial = axial_data[:, 1]

# Create the plot with settings from map_phase_diagram_improved.py
plt.figure(figsize=(12, 8))

# Plot the vector melting line first (background)
plt.plot(mu_vector, T_vector, color='red', linestyle='--', 
        label='Vector Melting Line', linewidth=2, alpha=0.8)

# Plot the axial melting line
plt.plot(mu_axial, T_axial, color='black', 
        label='Axial Melting Line', linewidth=2, alpha=0.8)

# Separate first order and crossover points
first_order = order == 1
crossover = order == 2

# Plot critical line with improved styling (matching map_phase_diagram_improved.py)
if np.any(first_order):
    plt.plot(mu_phase[first_order], Tc_phase[first_order], 'ro-', 
            label='First order transition', linewidth=3, markersize=8, alpha=0.9)

if np.any(crossover):
    plt.plot(mu_phase[crossover], Tc_phase[crossover], 'bo-', 
            label='Crossover/2nd order', linewidth=3, markersize=8, alpha=0.9)

# Labels and title (matching map_phase_diagram_improved.py)
plt.xlabel('Chemical Potential  (MeV)', fontsize=14)
plt.ylabel('Critical Temperature Tc (MeV)', fontsize=14)
plt.title(f'QCD Phase Diagram\n'
          f'$m_q = {mq}$ MeV, $\\lambda_1 = {lambda1_value:.3f}$', fontsize=16)

# Add grid and legend (matching map_phase_diagram_improved.py)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12, loc='best')

# Improve layout
plt.tight_layout()

# Create CP_plots directory if it doesn't exist
os.makedirs('CP_plots', exist_ok=True)

# Save the plot with improved naming (matching map_phase_diagram_improved.py)
plot_filename = f'CP_plots/phase_diagram_improved_ml{mq}_lambda1{lambda1_value}.png'
plt.savefig(plot_filename, dpi=300, bbox_inches='tight')
print(f"Plot saved as: {plot_filename}")

# Display the plot
plt.show()

# Print some statistics about the data
print(f"\nData summary:")
print(f"Chemical potential range: {mu_phase.min():.1f} - {mu_phase.max():.1f} MeV")
print(f"Critical temperature range: {Tc_phase.min():.1f} - {Tc_phase.max():.1f} MeV")
print(f"Number of phase transition points: {len(mu_phase)}")
print(f"First-order transitions: {np.sum(order == 1)}")
print(f"Second-order transitions: {np.sum(order == 2)}")
```
